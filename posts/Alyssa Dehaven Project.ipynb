```python
import pandas as pd
import matplotlib.pyplot as plt
import os

# Print to confirm the current working directory
print("Current Working Directory:", os.getcwd())

# Define the absolute path to the CSV file
file_path = '/home/jovyan/AlyssasBlog/posts/foster youth.csv'  # Adjust this path as needed

# Print the file path to verify
print("File Path:", file_path)
print("File Exists:", os.path.exists(file_path))

# Load the CSV file
df = pd.read_csv(file_path)

# Remove leading spaces from column names
df.columns = df.columns.str.strip()

# Convert relevant columns to numeric, ignoring errors for non-numeric entries
df['Total Matched'] = pd.to_numeric(df['Total Matched'], errors='coerce')

# Drop rows with NaN values in 'Total Matched' column
df = df.dropna(subset=['Total Matched'])

# Sort the dataframe by 'Total Matched' in descending order and select the top 3 counties
top_counties = df.nlargest(3, 'Total Matched')

# Plotting the data
plt.figure(figsize=(10, 6))
plt.bar(top_counties['County of Jurisdiction'], top_counties['Total Matched'], color=['blue', 'orange', 'green'])
plt.xlabel('County')
plt.ylabel('Number of Foster Youth')
plt.title('Top 3 Counties with the Highest Number of Foster Youth')
plt.xticks(rotation=45)
plt.show()
```
